```python
import pandas as pd
import altair as alt

data = pd.read_csv('./events.csv')
other = pd.read_csv('./athlete_events.csv')
```

```python
alt.data_transformers.disable_max_rows()
```

```python
hist = alt.Chart(data).mark_bar().transform_filter(
    (alt.datum.Sport == 'Gymnastics') & (alt.datum.Sex == 'F')
    & (alt.datum.Medal == "Gold")
).encode(
    x = alt.X('Height', title='Height (cm)'),
    y=alt.Y('count()', title='Count')).properties(
    title="Height Distribution of Women's Gymnastics Olympians Who Won a Gold Medal"
)

rule = alt.Chart(data).mark_rule(color = '#3ebede').transform_filter(
    (alt.datum.Sport == 'Gymnastics') & (alt.datum.Sex == 'F')
).encode(
    x = alt.X('mean(Height):Q', title="")
)
hist + rule
```
